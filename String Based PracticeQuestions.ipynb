```python
# 23. Check if a string is alphanumeric
def is_alphanumeric(string):
    return string.isalnum()

# 24. Extract all digits from a string
def extract_digits(string):
    return ''.join(char for char in string if char.isdigit())

# 25. Extract all alphabets from a string
def extract_alphabets(string):
    return ''.join(char for char in string if char.isalpha())

# 26. Count the number of uppercase letters in a string
def count_uppercase_letters(string):
    return sum(1 for char in string if char.isupper())

# 27. Count the number of lowercase letters in a string
def count_lowercase_letters(string):
    return sum(1 for char in string if char.islower())

# 28. Swap the case of each character in a string
def swap_case(string):
    return string.swapcase()

# 29. Remove a specific word from a string
def remove_word(string, word):
    return string.replace(word, '')

# 30. Check if a string is a valid email address
def is_valid_email(email):
    import re
    email_pattern = r"^\w+([\.-]?\w+)*@\w+([\.-]?\w+)*(\.\w{2,4})+$"
    return re.match(email_pattern, email) is not None

# 31. Extract the username from an email address string
def extract_username(email):
    return email.split('@')[0]

# 32. Extract the domain name from an email address string
def extract_domain_name(email):
    return email.split('@')[1]

# 33. Replace multiple spaces in a string with a single space
def replace_multiple_spaces(string):
    import re
    return re.sub(r'\s+', ' ', string)

# 34. Check if a string is a valid URL
def is_valid_url(url):
    import re
    url_pattern = r"^(http|https)://([\w-]+\.)+[\w-]+/([\w-]+\.)*[\w-]+(\?([\w-]+=[\w-]+&)*[\w-]+=[\w-]+)?$"
    return re.match(url_pattern, url) is not None

# 35. Extract the protocol (http or https) from a URL string
def extract_protocol(url):
    return url.split('://')[0]

# 36. Find the frequency of each character in a string
def character_frequency(string):
    from collections import Counter
    return Counter(string)

# 37. Remove all punctuation from a string
def remove_punctuation(string):
    import string
    return string.translate(str.maketrans('', '', string.punctuation))

# 38. Check if a string contains only digits
def contains_only_digits(string):
    return string.isdigit()

# 39. Check if a string contains only alphabets
def contains_only_alphabets(string):
    return string.isalpha()

# 40. Convert a string to a list of characters
def convert_to_list(string):
    return list(string)

# 41. Check if two strings are anagrams
def are_anagrams(string1, string2):
    from collections import Counter
    return Counter(string1) == Counter(string2)

# 42. Encode a string using a Caesar cipher
def caesar_cipher_encode(string, shift):
    import string
    alphabet = string.ascii_lowercase
    shifted_alphabet = alphabet[shift:] + alphabet[:shift]
    table = str.maketrans(alphabet, shifted_alphabet)
    return string.translate(table)

# 43. Decode a Caesar cipher encoded string
def caesar_cipher_decode(string, shift):
    import string
    alphabet = string.ascii_lowercase
    shifted_alphabet = alphabet[-shift:] + alphabet[:-shift]
    table = str.maketrans(shifted_alphabet, alphabet)
    return string.translate(table)

# 44. Find the most frequent word in a string
def most_frequent_word(string):
    from collections import Counter
    words = string.split()
    word_counts = Counter(words)
    return word_counts.most_common(1)[0][0]

# 45. Find all unique words in a string
def unique_words(string):
    words = string.split()
    return list(set(words))

# 46. Count the number of syllables in a string
def count_syllables(string):
    # A more accurate algorithm would consider vowel patterns and exceptions
    return len(string.split())

# 47. Check if a string contains any special characters
def contains_special_characters(string):
    import string
    return any(char not in string.ascii_letters + string.digits + string.whitespace for char in string)

# 48. Remove the nth word from a string
def remove_nth_word(string, n):
    words = string.split()
    if n < 1 or n > len(words):
        return string
    words.pop(n - 1)
    return ' '.join(words)

# 49. Insert a word at the nth position in a string
def insert_word(string, word, n):
    words = string.split()
    if n < 1 or n > len(words) + 1:
        return string
    words.insert(n - 1, word)
    return ' '.join(words)

# 50. Convert a CSV string to a list of lists
def convert_csv_to_list(csv_string):
    import csv
    csv_string = csv_string.replace('\n', '')
    reader = csv.reader([csv_string])
    return list(reader)[0]
```
